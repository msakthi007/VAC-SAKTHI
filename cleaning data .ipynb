```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import Normalizer
```

```python
normalization_df=pd.DataFrame({'x':np.random.randint(-100,100,1000),'y':np.random.randint(-80,80,1000),'z':np.random.randint(20,40,1000)})
```

```python
normalization_df
```

```python
from numpy.random.mtrand import normal
ax=plt.axes(projection='3d')

ax.scatter3D(normalization_df.x,normalization_df.y,
             normalization_df.z)
plt.figure()
```

```python
nol=Normalizer()
```

```python
normalization_df=nol.fit_transform(normalization_df)
```

```python
normalization_df
```

```python
norm=pd.DataFrame(normalization_df,columns=["x","y","z"])

```

```python
ax=plt.axes(projection='3d')

ax.scatter3D(norm.x,norm.y,
          norm.z)
plt.figure()
```
